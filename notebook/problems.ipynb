## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np
```

```python
# TODO

def dice_prob(rolls):
    
    # Create a counter to sum the number of times the result is 7 or even number. 
    counter = 0 

    # Roll the dice 'rolls' times and do a sum of the result  
    for i in range(rolls):
        d1 = np.random.randint(1,7)
        d2 = np.random.randint(1,7)
        dice_sum = d1 + d2

    # If dice_sum is greater than 7 or an even number, then add 1 to our counter        
        if dice_sum > 6 or dice_sum % 2 == 0:
            counter += 1

    # Calculate the probability 
    prob = counter / rolls  
    # Print the probability
    print('The probability in {} rolls that the dice is greater than 7 or an even number is {}'.format(rolls, prob))
```

```python
dice_prob(1000)
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
def prob_colors(iterations = 1000):
    counter1 = 0 # Counter for the 3 white and 2 red
    counter2 = 0 # Counter for all the same color 
    
    # Total number of iterations
    for i in range(iterations):
    
        # Reseting the colors 
        white = 0 
        red = 0 
        green = 0 
    
        # Picking 5 balls and saving the colors (first 10 are white, the next 20 are red and the last 30 are green)
        for i in range(5):
            bpick = np.random.randint(1,61)
            if bpick < 11:
                white += 1
            elif bpick < 31:
                red += 1
            else:
                green += 1
    
        # Counting how many picks were 3 white and 2 red
        if (white == 3) & (red == 2):
            counter1 += 1
        # Counting how many picks where all the same color 
        if (white == 5) or (red == 5) or (green == 5):
            counter2 += 1
    
    print('The probability of getting 3 white and 2 red is {}%'.format(round((counter1/iterations),3)*100))
    print('The probability of getting all the same color is {}%'.format(round((counter2/iterations),3)*100))    
```
